```python
biglist1=[]
biglist2=[]
biglist3=[]
for j in range(70):
    list1=[]
    list2=[]
    list3=[]
    for i in range(file2json[j]['results_shown']):
        a = file2json[j]['restaurants'][i]['restaurant']['user_rating']['aggregate_rating']
#     print(a)
        b = file2json[j]['restaurants'][i]['restaurant']['user_rating']['votes'] 
#     print(b) 
        q=float(a)*float(b)
        list1.append(q)
        list2.append(file2json[j]['restaurants'][i]['restaurant']['name'])
        cityname=(file2json[j]['restaurants'][i]['restaurant']['location'])['city']
        list3.append(cityname)
    biglist1.extend(list1)
    biglist2.extend(list2)
    biglist3.extend(list3)
```

```python
bigtuple=(zip(biglist1,biglist2,biglist3))
print(len(biglist1))
```

```python
sortedlist=sorted(bigtuple, reverse=True)
```

```python
print(sortedlist)
```

```python
sortedlist2=[]
for i in range(len(sortedlist)):
    if(sortedlist[i] not in sortedlist2):
        sortedlist2.append(sortedlist[i]) 
```
